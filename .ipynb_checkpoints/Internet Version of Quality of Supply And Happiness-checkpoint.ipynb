**tl;rd I've conducted a whole bunch of data analysis trying to find support for Thoughty2's claim that scandinavians rig the results of happiness surveys. Turns out they do a little bit.**

(Reading time ~8 min)

For those who don't know what I'm talking about, watch [this youtube video](https://youtu.be/tckVx_kaYco)

Basically Thoughty2's argument is, that this culture <s>forces</s> encourages people to feel happier than they would be otherwise, given their circumstances in life. His second point is that [Janteloven](https://en.wikipedia.org/wiki/Law_of_Jante) would encourage folks to report higher happiness numbers, higher even then they actually think the answer is.

I decided I can help shed some light onto this hypothesis. I found some data about what people in each european country THINK the quality of electric supply is in their country, and some data about how good it ACTUALLY is.. And I set out to see if there is any evidence that people in <s>scandinavian</s> certain countries tend to overestimate how good things in their life are.  Later I compare attained biases with the biases from World Happiness Report

So here's the data:

"poll" column : results of the survey where the question was: "In your country, how reliable is the electricity supply (lack of interruptions and lack of voltage fluctuations)? [1 = extremely unreliable; 7 = extremelyreliable]"

"saidi" column - actual number of minutes per customer per year that there was no electricity

*Note: SAIDI - System Average Interruption Duration Index*

Here is a scatterplot of the two columns

Each dot represents one country's "poll" and "saidi" results. The dotted black line is a linear approximation of the data.

As you might have noticed all of the orange dots (Scandinavian countries) lie above the linear approximation. This means these countries when surveyed are overly positive about estimating how good their electrical supply is.
All of the Scandinavian countries always lie above linear approximation line in the data for the years 2012-2016.

*Note: there's no data on Iceland...sorry Iceland*

On the barplot below you can see how much each of the countries is deviating from it's predicted by linear model "poll" score. This is for year 2016

### Let's now look at the results of **World Happiness Report**

Here are top 10 happiest nations according to 2015 survey

As you see Scandi countries rank pretty high

Here's the trickiest part to understand. The people who publish this survey and who are much smarter than me, alongside with rankings, were recording a whole bunch of other data, such as income, healthy life expectancy, social support, freedom, trust and generosity. Each country was given a score on each of these 6 measurements. All of them have been found to affect happiness. In fact they have been used to try to predict the results of happiness scores.

The models they have built are not perfect and happiness can't be calculated based on just these 6 measurements.
That's why for each country there is some additional amount of happiness that can't be explained by their models.

I decided to record this unexplained amount of happiness in a columns called "positivity". So if the country when surveyed scores higher than it's predicted happiness, then the value in "positivity" column will be positive. If lower, it will be negative. ( I could not put enough asterisks on that explanation, for more info click [here](https://nbviewer.jupyter.org/github/SerjTim/Quality_Of_Supply_And_Happiness/blob/master/Analysis.ipynb))

That number I compared with biases that I got from quality of supply research (the barplot above). That I call "electrical positivity" or "el positivity" in short. That is how much people over or under estimated their country's electric supply quality.
You can see the numbers in the table below

**And now, finally we get to the interesting part**

You can observe how biases that come from two completely separate places (electric quality survey and world happiness survey) are actually highly correlated

In fact the correlation coefficient is 67%

Because these two questions (quality of supply and life happiness) have pretty much nothing to do with each other, the fact that people's biases answering them are this much correlated, points to the idea that big part of the reason for these biases is cultural (as opposed to being just a rondom noise or a measuring error). As for Scandinavian countries, Janteloven seems to be a valid explanation for the cultural difference, but until further research it is as good an explanation as any.

**To conclude:** for Scandinavian countries, while the above evidence supports the hypothesis that they are overly positive reporting the happiness results, it is no doubt that life in those countries is objectively good. Even without that positivity bias they would be somewhere in the top half of the happiness rankings
